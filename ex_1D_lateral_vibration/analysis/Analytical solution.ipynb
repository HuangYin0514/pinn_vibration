# function

```python
def get_analytical_solution(t, x):
    u = np.sin(np.pi * x) * np.cos(np.pi**2 * t)
    return u
```

# infer

```python
x = np.linspace(0, 1, 256)
t = np.linspace(0, 1, 100)
X, T = np.meshgrid(x, t)
analytical_solution = get_analytical_solution(T, X)
analytical_solution.shape
```
